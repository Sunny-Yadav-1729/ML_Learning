```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
```

```python

df = pd.read_csv(r"e:\dist\boston_house_prices_1_1.csv",header=[1])
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import test


x= np.array(df.iloc[:,:-1])
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
lr = LinearRegression()
lr.fit(x_train,y_train)
r2_score(y_test,lr.predict(x_test))
sc =StandardScaler()
scc =StandardScaler()
# sc.fit(x)
x = sc.fit_transform(x)
# y = sc.fit_transform(y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor


esti = [
    LinearRegression(),
    DecisionTreeRegressor(random_state=42),
    KNeighborsRegressor(),
    SVR()
]
```

```python
from sklearn.model_selection import cross_val_score


for m in esti:
    # m = i
    
    s = cross_val_score(m,x,y,cv=2,scoring='r2')
    print(m,np.round(np.mean(s),2))
```

```python
from sklearn.ensemble import BaggingRegressor


```

```python
n_sam = [10,20,50,60]
for m in esti:
    for n in n_sam:
        bag = BaggingRegressor(m,n)
        print(m,n,cross_val_score(bag,x,y,cv=5,scoring='r2').mean())
```

```python
from sklearn.model_selection import GridSearchCV

param =  { 
        'estimator':esti,     
        'n_estimators':n_sam,
        "max_samples" :[.5,1],
        'bootstrap' :[True,False]    
     }

bag_GSCV = GridSearchCV(BaggingRegressor(random_state=1,n_jobs=-1),param_grid=param,cv=3,verbose=1,n_jobs=-1)
```

```python
bag_GSCV.fit(x,y)
```

```python

print('Test R^2 Score : %.3f'%bag_GSCV.best_estimator_.score(x, y))
print('Best R^2 Score Through Grid Search : %.3f'%bag_GSCV.best_score_)
print('Best Parameters : ',bag_GSCV.best_params_)
```
