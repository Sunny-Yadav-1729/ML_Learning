```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\WELCOME\OneDrive\Desktop\Learning\Enemble_Learning\classification__datasets\xor_dataset.csv")
```

```python
x= np.array(df.iloc[:,:2])
y = df.iloc[:,-1]
```

```python
from sklearn.ensemble import RandomForestClassifier


RFC = RandomForestClassifier()
```

```python
from sklearn.model_selection import cross_val_score


cross_val_score(RFC,x,y,cv=5,scoring='accuracy').mean()
```

```python
def get_RFC_pred(n,max_dept,s):
    RFC = RandomForestClassifier(n,max_depth=max_dept,max_samples=s)
    a = cross_val_score(RFC,x,y,cv=5,scoring='accuracy').mean()
    print(n,max_dept,s," - >",a)
```

```python
for i in range(50,501,50):

    get_RFC_pred(i,i,np.random.rand())
```
