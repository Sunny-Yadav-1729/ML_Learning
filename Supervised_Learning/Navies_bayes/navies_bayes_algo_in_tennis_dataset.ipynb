```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\WELCOME\OneDrive\Desktop\Learning\classification__datasets\play_tennis.csv').iloc[:,1:]
df 
```

```python
Look_up_Table = {}
df.shape[0]
x = df.iloc[:,:-1]
y = df.iloc[:,-1:]
x.select_dtypes(include=['object', 'category']).columns
```

```python
sum(df[df['play']=='Yes']['outlook'] == 'Rain')/len(df[df['play']=='Yes'])
```

```python
Look_up_Table['play'] = {}
for i in y['play'].unique():
    p = len(y[y['play']==i])/y.shape[0]
    Look_up_Table['play'][i] = p
col_name = x.select_dtypes(include=['object', 'category']).columns
for c in col_name:
    for feat in df[c].unique():
        for n_f in df['play'].unique():
            p = sum(df[df['play']==n_f][c] == feat)/len(df[df['play']==n_f])
            if c not in Look_up_Table:
                Look_up_Table[c] = {}
            if feat not in Look_up_Table[c]:
                Look_up_Table[c][feat] = {}
            if n_f not in Look_up_Table[c][feat]:
                Look_up_Table[c][feat][n_f] = p



           
    



def predict(x):
    l = {}
    l['Yes'] = Look_up_Table['play']['Yes']
    l['No'] = Look_up_Table['play']['No']
 
   
    for c in x.columns:
       for f in x[c].unique():
           for b in ['Yes','No']:
               l[b]*= Look_up_Table[c][f][b] 
    ans = 'Yes'
    val = -1
    for k,v in l.items():
        if val == -1 or v > val:
            ans = k
            val = v
    return {ans,val}
```

```python
predict(x.iloc[:1])
```

```python
def get(n):
    print(predict(x.iloc[n-1:n]))
```
