```python
import pandas as pd
import numpy as np
```

```python
from sklearn.ensemble import StackingClassifier, StackingRegressor


```

```python
df = pd.read_csv((r"C:\Users\WELCOME\OneDrive\Desktop\Learning\classification__datasets\xor_dataset.csv"))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC


st = StackingClassifier([
    ('lr',LogisticRegression()),
    ('lrs',SVC()),
])
```

```python
x = df.iloc[:,:2]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import cross_val_score


cross_val_score(st,x,y,cv=10,scoring='accuracy').mean()
```
